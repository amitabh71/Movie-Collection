```python
from sklearn.model_selection import train_test_split

X_train_full , X_test , y_train_full , y_test =train_test_split(movie_data ,movie_target_data , random_state=42)
X_train ,X_valid ,y_train , y_valid=train_test_split(X_train_full,y_train_full, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_valid=scaler.transform(X_valid)
X_test=scaler.transform(X_test)
```

```python
import numpy as np
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
X_valid.shape
```

```python
X_test.shape
```

```python
model =keras.models.Sequential(
    [keras.layers.Dense(30,activation="relu",input_shape=[19]),
     keras.layers.Dense(30,activation="relu"),
     keras.layers.Dense(1)
     
        
    ]
)
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",
             optimizer=keras.optimizers.SGD(lr=1e-3),
             metrics=['mae'])
```

```python
model_history=model.fit(X_train ,y_train,epochs=50,validation_data=(X_valid,y_valid))
```

```python
mae_test=model.evaluate(X_test ,y_test)
```
